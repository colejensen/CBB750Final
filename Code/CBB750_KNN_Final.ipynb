```python
# import the needed packages
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# read in the data
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/dataset.xlsx', index_col=False)
# make sure it looks right
print(df.shape)
df.head()
```

```python
from sklearn import preprocessing

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'species'.
df['SARS-Cov-2 exam result']= label_encoder.fit_transform(df['SARS-Cov-2 exam result'])
 
df['SARS-Cov-2 exam result'].unique()
```

```python
df.head()
```

```python
# now we need to deal with the missing values 
df.columns = [x.lower().strip().replace(' ','_') for x in df.columns]
```

```python
def miss_data(data_frame):
    total = data_frame.isnull().sum()
    percent = (data_frame.isnull().sum()/data_frame.isnull().count()*100)
    missing_df = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    types = []
    for col in data_frame.columns:
        dtype = str(data_frame[col].dtype)
        types.append(dtype)
    missing_df['Types'] = types
    return(np.transpose(missing_df))
```

```python
miss_data(df)
```

```python
for i in df.columns:
    if df[i].dtype=='float16' or  df[i].dtype=='float32' or  df[i].dtype=='float64':
        df[i].fillna(df[i].mean())

df = df.fillna(-999)

for j in df.columns:
    if df[j].dtype=='object': 
        label = preprocessing.LabelEncoder()
        label.fit(list(df[j].values))
        df[j] = label.transform(list(df[j].values))
```

```python
threshold = 0.90

corr_matrix = df.corr().abs()

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
upper.head()
```

```python
# drop the missing columns 
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

print('There are %d columns to remove.' % (len(to_drop)))
dataset = df.drop(columns = to_drop)
```

```python
missing_df = (df.isnull().sum() / len(df)).sort_values(ascending = False)
missing_df.head()
```

```python
df.isnull().sum()
```

```python
df.replace({-np.inf: -1_000_000, np.inf: 1_000_000}, inplace=True)
```

```python
missing_df = missing_df.index[missing_df > 0.85]
all_missing = list(set(missing_df))


dataset = dataset.drop(columns = all_missing)
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
cols = [x for x in dataset.columns if x not in ['patient_id']]
new_df = dataset[cols]
new_df
```

```python
hospital_status = []
for i in range(0, len(new_df)):
  if new_df['patient_addmited_to_regular_ward_(1=yes,_0=no)'][i] == 1:
    to_append = 1
  elif new_df['patient_addmited_to_semi-intensive_unit_(1=yes,_0=no)'][i] == 1:
    to_append = 2
  elif new_df['patient_addmited_to_intensive_care_unit_(1=yes,_0=no)'][i] == 1:
    to_append = 3
  else:
    to_append = 0
  hospital_status.append(to_append)

hospital_status = pd.DataFrame(hospital_status)
hospital_status = hospital_status.rename(columns={hospital_status.columns[0]: "Hospital Treatment Type"})
new_df = pd.concat([new_df, hospital_status], axis=1)
```

```python
# these variables are no longer needed so remove them now 
# dropping the selected columns 
new_df.drop(columns=['patient_addmited_to_regular_ward_(1=yes,_0=no)',
                      'patient_addmited_to_semi-intensive_unit_(1=yes,_0=no)',
                      'patient_addmited_to_intensive_care_unit_(1=yes,_0=no)'], inplace=True)
new_df.head()
```

```python
new_df.to_csv("data_to_use.csv", index=False)
!cp data_to_use.csv "drive/My Drive/Colab Notebooks/"
```

```python
# import the KNeighborsClassifier module
from sklearn.neighbors import KNeighborsClassifier

# instantiating KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
# for splitting data into training and testing data
from sklearn.model_selection import train_test_split

# defining target variables 
target = new_df['Hospital Treatment Type']

# defining predictor variables 
features = new_df.select_dtypes(exclude=[object])

# assigning the splitting of data into respective variables
x_train,x_test,y_train,y_test = train_test_split(features, target, test_size=0.3, random_state=42, stratify = target)
```

```python
# get some stats about the different sets
print("Number of samples in train set: %d" % y_train.shape)
print("Number of positive samples in train set: %d" % (y_train == 1).sum(axis=0))
print("Number of negative samples in train set: %d" % (y_train == 0).sum(axis=0))
print()
print("Number of samples in test set: %d" % y_test.shape)
print("Number of positive samples in test set: %d" % (y_test == 1).sum(axis=0))
print("Number of negative samples in test set: %d" % (y_test == 0).sum(axis=0))
```

```python
# to display the HTML representation of an object.
from IPython.display import display_html

x_train_data = x_train.describe().style.set_table_attributes("style='display:inline'").set_caption('Summary of Training Data')
x_test_data = x_test.describe().style.set_table_attributes("style='display:inline'").set_caption('Summary of Testing Data')

# to display the summary of both training and testing data, side by side for comparison 
display_html(x_train_data._repr_html_(), raw = True)
display_html(x_test_data._repr_html_(), raw = True)
```

```python
# for exhaustive search over specified parameter values for an estimator
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors' : np.arange(1,40)}
```

```python
# performing Grid Search CV on knn-model, using 7-cross folds for validation of each criteria
knn_cv = GridSearchCV(knn, param_grid, cv=7)
```

```python
# training the model with the training data and best parameter
knn_cv.fit(x_train,y_train)
```

```python
# finding out the best parameter chosen to train the model
print("The best paramter we have is: {}" .format(knn_cv.best_params_))

# finding out the best score the chosen parameter achieved
print("The best score we have achieved is: {}" .format(knn_cv.best_score_))
```

```python
# get the predictions
y_pred = knn_cv.predict(x_test)
```

```python
# Calculate the mean accuracy for the test data
print("The score accuracy for training data is: {}" .format(knn_cv.score(x_train,y_train)))
print("The score accuracy for testing data is: {}" .format(knn_cv.score(x_test,y_test)))
```

```python
# for performance metrics
from sklearn.metrics import classification_report, confusion_matrix, plot_confusion_matrix

# call the classification_report and print the report
print(classification_report(y_test, y_pred))
```

```python
# call the confusion_matrix and print the matrix
print(confusion_matrix(y_test, y_pred))
```
